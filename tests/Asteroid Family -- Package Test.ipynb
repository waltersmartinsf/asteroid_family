```python
%matplotlib inline
import asteroid_family as family
import numpy as np
import matplotlib.pyplot as plt
import seaborn
plt.rcParams['figure.figsize'] = (14.0,8.0) # change figure size
from pandas import *
import os #gerenciamento sistema operacional
import astropy.io.misc.pickle_helpers as astrobin #criacao do binario python
```

```python
#Parental data
rpb = 20 #km

rho_mantle = 3.
rho_core = 7.

pv_mantle = 0.4
pv_core = 0.15

Vi = 5 #km/s impact velocity
fke = 0.01 #inelastic

#298 Baptistina
M       = 280.1642437983165 #degrees
peri    = 135.1543076468920 #degrees
node    = 8.253490030039535 #degrees
i       = 6.28752726137564 #degrees
e       = 0.09612545504215386
a       = 2.26362004764974 #AU
period  = 3.41 #year
ano     = 2013
mes     = 11
dia     = 4
da0     = 0
```

```python
massa, vej, rho, raio, mag = family.differentiated_family('H',rpb,rho_mantle,rho_core,Vi,fke,0.5,pv_core,pv_mantle,'YES')
VT, VR, VW, A, E, I, dA, dE, dI = family.gauss_equations(vej, a, e, i, period, 'YES')
dadt = family.yarkovsky_dadt(2*raio,np.random.uniform(0,np.pi,len(massa))*180./np.pi)
```

```python
data = DataFrame([massa, vej, rho, raio, mag,VT, VR, VW, A, E, I, dA, dE, dI, dadt]).T
data.columns = ['massa', 'vej', 'densidade', 'raio', 'H', 'VT', 'VR', 'VW', 'A', 'E', 'I', 'dA', 'dE', 'dI', 'dadt']
data
```

```python
plt.figure()
plt.grid(color='white')
plt.scatter(data.A[data.densidade == 3],data.E[data.densidade == 3], color='blue', label='mantle = '+str(len(data.E[data.densidade == 3])))
plt.scatter(data.A[data.densidade == 7],data.E[data.densidade == 7], color='red', label='core = '+str(len(data.E[data.densidade == 7])))
plt.legend()
plt.xlabel('a [AU]', fontsize=20)
plt.ylabel('e', fontsize=20)
```

```python
plt.figure()
plt.grid(color='white')
plt.scatter(data.raio[data.densidade == 3],data.dadt[data.densidade == 3], color='blue', label='mantle = '+str(len(data.E[data.densidade == 3])))
plt.scatter(data.raio[data.densidade == 7],data.dadt[data.densidade == 7], color='red', label='core = '+str(len(data.E[data.densidade == 7])))
plt.legend()
plt.xlabel('raio [km]', fontsize=20)
plt.ylabel('da/dt [AU/Myr]', fontsize=20)
plt.ylim(min(data.dadt),max(data.dadt))
```
